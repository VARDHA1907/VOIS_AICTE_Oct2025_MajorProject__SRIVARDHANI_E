```python
#Installing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the csv file and renaming Netflix Dataset to Netflix_Dataset for white space avoidance
df = pd.read_csv('Netflix_Dataset.csv', encoding='utf-8')
```

```python
#Check the first five rows and columns
print(df.head())
```

```python
df.head()
```

```python
# Convert Release_Date to datetime format and extract the year
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['Year'] = df['Release_Date'].dt.year
```

```python
df.head()
```

We can notice that we have a seperate column namely year which can be used for time based analysis


```python
#changing year format to int
df['Year'] = df['Year'].astype('Int64')
```

```python
#checking whether year column changed to int
df.info()
```

```python
 #Standardize text columns, handle whitespace, and prepare exploded columns
text_cols = ['Category','Title','Director','Cast','Country','Rating','Duration','Type','Description']
for c in text_cols:
    if c in df.columns:
        df[c] = df[c].astype(str).str.strip().replace({'': np.nan})
```

```python
# Split multi-label columns (Type, Country) safely later; create helpers now
df['Type_list'] = df['Type'].fillna('').str.split(',').apply(lambda x: [t.strip() for t in x if t.strip() != ''])
df['Country_list'] = df['Country'].fillna('').str.split(',').apply(lambda x: [t.strip() for t in x if t.strip() != ''])
```

```python
df.head()
```

We can notice that :
1. White space are removed
2. Multi column values are created with helper for now
3. And Year is changed to int


```python
# Duration parsing: minutes for Movies, seasons for TV Shows
def parse_duration(row):
    val = str(row['Duration']) if pd.notna(row['Duration']) else ''
    if 'min' in val:
        try:
            return int(val.split(' ')[0])
        except:
            return np.nan
    if 'Season' in val:
        try:
            return int(val.split(' ')[0])
        except:
            return np.nan
    return np.nan
```

**Duration Parsing Logic**

Extracts numeric value from Duration column.

Handles both "min" for Movies and "Season" for TV Shows.

Invalid or missing entries are safely replaced with np.nan.

```python
# Create helper columns for duration and content type flags
df['Duration_value'] = df.apply(parse_duration, axis=1)
df['Is_TV'] = (df['Category'] == 'TV Show').astype(int)
df['Is_Movie'] = (df['Category'] == 'Movie').astype(int)
```

Helper Columns for Analysis

Duration_value: Parsed numeric duration (minutes or seasons).

Is_TV and Is_Movie: Binary flags for content type, useful for filtering and aggregation.

```python
print(df[['Category','Duration','Duration_value','Year']].head())
print('Cleaned text fields, parsed duration, and created helper flags')
```

1. Finally we have the cleaned text fields, parsed duration and creating helper flags.

```python
# Annual distribution of Movies vs TV Shows
annual = df.dropna(subset=['Year']).groupby(['Year','Category']).size().reset_index(name='Count')
pivot_annual = annual.pivot(index='Year', columns='Category', values='Count').fillna(0)
```

```python
print(pivot_annual.head())
```

```python
# Line plot: Annual count of Movies vs TV Shows
plt.figure(figsize=(10,5))
pivot_annual[['Movie','TV Show']].plot(kind='line', marker='o')
plt.title('Movies vs TV Shows by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print('Plotted Movies vs TV Shows by Year')
```

Trend Analysis: Movies vs TV Shows (2008–2021)
Growth Phase (2015–2019): Both categories saw a sharp rise in production starting around 2015.

1. Movies peaked in 2019 with over 1400 releases, indicating a surge in film output.

2. TV Shows grew steadily, peaking in 2020 with around 700 releases, reflecting the streaming boom.

3. Early Years (2008–2014): Content volume was minimal, suggesting limited platform reach or catalog availability.

4. Recent Decline (2020): Both categories show a drop, possibly due to pandemic-related production delays or strategic shifts in content planning.

```python
# Normalize to show shares by year
share = pivot_annual.div(pivot_annual.sum(axis=1), axis=0)

print(share.head())
```

```python
# Area plot: Share of Movies vs TV Shows over time
plt.figure(figsize=(10,5))
share[['Movie','TV Show']].plot(kind='area', stacked=True, alpha=0.6)
plt.title('Share of Movies vs TV Shows Over Time')
plt.xlabel('Year')
plt.ylabel('Share')
plt.tight_layout()
plt.show()

print('Plotted share of content types over time')
```

**Content Share Dynamics (2008–2020)**

Movie Dominance (2008–2012) Movies consistently held the majority share, often exceeding 90% of total content. TV Shows were minimal during this period, indicating limited catalog depth or platform focus.

Shift Begins (2013–2015) A noticeable dip in movie share around 2013 marks the beginning of a strategic shift. TV Shows start gaining traction, possibly due to the rise of serialized streaming content.

Streaming Boom (2016–2020) TV Shows steadily increase their share, peaking around 2020. This reflects growing investment in episodic content and global demand for binge-worthy series.

International Movies lead by a wide margin (~2300 titles), reflecting Netflix’s global content strategy.

International TV Shows, TV Dramas, and TV Comedies feature prominently, indicating a well-rounded episodic catalog.

```python
# Indicators to support strategic recommendations
latest_year = int(df['Year'].dropna().max()) if df['Year'].notna().any() else None
```

```python
# Last 3 years mix
last3 = df[df['Year'].between(latest_year-2, latest_year, inclusive='both')] if latest_year is not None else df.copy()
mix_last3 = last3['Category'].value_counts(normalize=True) if len(last3) else pd.Series(dtype=float)
```

```python
# Momentum for top genres: CAGR-like ratio of last3 avg vs first3 avg
types_year_full = (
    df.explode('Type_list')
      .dropna(subset=['Year', 'Type_list'])
      .groupby(['Year', 'Type_list'])
      .size()
      .reset_index(name='Count')
)
```

```python
if not types_year_full.empty:
    yrmin = int(types_year_full['Year'].min())
    yrmax = int(types_year_full['Year'].max())

    # First 3 years average
    first3_avg = (
        types_year_full[types_year_full['Year'].between(yrmin, min(yrmin + 2, yrmax))]
        .groupby('Type_list')['Count']
        .mean()
    )

    # Last 3 years average
    last3_avg = (
        types_year_full[types_year_full['Year'].between(max(yrmax - 2, yrmin), yrmax)]
        .groupby('Type_list')['Count']
        .mean()
    )
     # Momentum ratio
    momentum = (
        (last3_avg / first3_avg)
        .replace([np.inf, -np.inf], np.nan)
        .dropna()
        .sort_values(ascending=False)
        .head(10)
    )
else:
    momentum = pd.Series(dtype=float)

# Display result
print(momentum)
```
